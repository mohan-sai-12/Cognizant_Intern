### IMPORTING LIBRARIES 

```python
import spacy
from spacy import displacy
import pandas as pd
from urllib.parse import quote
```

```python
nlp = spacy.load("en_core_web_sm")
file=open('NLPPipeline.txt',encoding="mbcs").read()
of=nlp(file)
```

### STEP 1:- Sentence Segmentation

```python
sent=of.sents
print(sent)
```

```python
for se in sent:
    print(se)
```

```python
sent1=list(of.sents)
print(sent1)
print(len(sent1))
```

### STEP 2 :- Word Tokenization

```python
print([token.text for token in of])
```

### STEP 3:- Predicting Parts Of Speech for each token

```python
l=[]
for token in of:
    r=token.text,token.pos_,token.tag_,spacy.explain(token.pos_),spacy.explain(token.tag_)
    l.append(r)
df=pd.DataFrame(l)
print(df.head(4))
```

### STEP 4:- Text Lemmatization

```python
l1=[]
sen=[]
for token in of:
    r=token.text,token.lemma_
    l1.append(r)
    sen.append(r[1])
sen=' '.join(sen)
df=pd.DataFrame(l1,columns=['Token','Lemma'])
print(df.head(5))
print()
print(sen)
```

### STEP :-5 Remove Stop Words

```python
twsw=[]
stopwords = nlp.Defaults.stop_words
#print(stopwords)

for token in nlp(sen):
    if token.text not in stopwords:
        twsw.append(token.text)
print(twsw)
```

```python
sentence=" ".join(twsw)
print(sentence)
```

### STEP :-6a Dependency Parsing

```python
tokens = nlp(sen)
```

```python
displacy.render(tokens, style='dep', jupyter=True, options={'distance':100})
```

### STEP :-6b Finding Noun Phrases

```python
for np in tokens.noun_chunks:
    print(np.text)
```

### STEP :-7 Named Entity Recognition (NER)

```python
displacy.render(tokens, style="ent")
```
